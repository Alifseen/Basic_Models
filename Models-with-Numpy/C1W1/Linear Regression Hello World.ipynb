```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1.0, 2.0])
y_train = np.array([300, 500])
```

```python
m = x_train.shape[0]
m
```

```python
i = 0

x_i = x_train[i]
y_i = y_train[i]

print(x_i, y_i)
```

```python
plt.scatter(x_train, y_train)
plt.show()
```

```python
w=100
b=100
```

```python
def compute_output(x, w, b):
    m=x.shape[0]
    fx = np.zeros(m)
    for i in range(m):
        fx[i] = w*x[i]+b
    return fx
```

```python
y_hat = compute_output(x_train, w, b)
y_hat
```

```python
plt.plot(x_train, y_hat)
plt.scatter(x_train, y_train)
```

```python
plt.plot(x_train, compute_output(x_train, 200, 100))
plt.scatter(x_train, y_train)
```

```python
compute_output(np.array([1.2]), 200, 100)
```
