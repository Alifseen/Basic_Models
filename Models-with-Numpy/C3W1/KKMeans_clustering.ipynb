```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def find_closest_centroids(x, centroids):
    K = centroids.shape[0]

    idx = np.zeros(x.shape[0], dtype=int)

    for i in range(len(idx)):
        c = []
        for k in range(K):
            c_i = np.linalg.norm(x[i] - centroids[k])
            c.append(c_i)
        idx[i] = np.argmin(c)
                
    return idx
```

```python
xtr = np.load('data/ex7_X.npy')
```

```python
initial_centroids = np.array([[3,3], [6,2], [8,5]])

# Find closest centroids using initial_centroids
idx = find_closest_centroids(xtr, initial_centroids)

# Print closest centroids for the first three elements
print("First three elements in idx are:", idx[:3])
```

```python
def compute_centroids(x, idx, k):
    centroids = np.zeros((k, x.shape[1]))
    for j in range(k):
        c_k = x[idx==j]
        centroids[j] = np.mean(c_k , axis=0)
    return centroids
```

```python
K = 3
centroids = compute_centroids(xtr, idx, K)

print("The centroids are:", centroids)
```

```python
def run_Kmeans(x, Init_centroids, iters=10):
    idx = np.zeros(x.shape[0])
    centroids = Init_centroids
    for i in range(iters):
        print("K-Means iteration %d/%d" % (i, iters-1))
        idx = find_closest_centroids(x, centroids)
        centroids = compute_centroids(x, idx, centroids.shape[0])
    return centroids, idx
```

```python
def random_init_centroids(x, k):
    rndidx = np.random.permutation(x.shape[0])
    centroids = x[rndidx[:k]]
    return centroids
```

```python
initial_centroids = random_init_centroids(xtr, 3)

centroids, idx = run_Kmeans(xtr, initial_centroids)
```
